```python
cols = ['MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection',
        'TechSupport','StreamingTV','StreamingMovies']

for col in cols:
    df[col] = df[col].astype(str).str.strip().str.lower()
    df[col] = df[col].replace('no internet service', 'No')
    df[col] = df[col].replace('no phone service', 'No')
```

```python
df['Churn_flag'] = df['Churn'].map({'Yes': 1, 'No': 0}) 
```

```python
df
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
overall_Churn=df['Churn_flag'].mean()*100
print(f"overall_Churn Rate:{overall_Churn:.2f}%")
```

```python
corr=df[['tenure','MonthlyCharges','TotalCharges','Churn_flag']].corr()
plt.figure(figsize=(5,4))
plt.imshow(corr,cmap='coolwarm',interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr.columns)),corr.columns,rotation=45)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.title("Correlation HeatMap")
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
binary_cols = ['gender','Partner','Dependents','PhoneService','MultipleLines',
               'OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport',
               'StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for i in binary_cols:
    df[i]=le.fit_transform(df[i])
```

```python
df = pd.get_dummies(df,columns=['InternetService','Contract','PaymentMethod'],drop_first=True)
```

```python
x = df.drop(['customerID','Churn','Churn_flag'],axis=1)
x
```

```python
x.isnull().sum()
```

```python
x=x.fillna(0)
```

```python
x.isnull().sum().sum()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
df['Segment'] = kmeans.fit_predict(X_scaled)
```
